#### Notes
1) Haarcascade files can detect single or multiple occurences of an object from an image of varying shapes and sizes.

2) <b>detectMultiScale()</b> - Detects single or multiple occurrence of an object of different sizes in the input image. The detetced features are returned as coordinates.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
face_cascade = cv2.CascadeClassifier("E:\\Img_datasets\\haarcascade_files\\haarcascade_frontalface_alt.xml")
eye_cascade = cv2.CascadeClassifier("E:\\Img_datasets\\haarcascade_files\\haarcascade_eye.xml")
```

#### Face Detection  

```python
def face_detect(img):
    faces = face_cascade.detectMultiScale(img)
    for x,y,w,h in faces:
        
        img = cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),3)
    print(x,y,w,h)
    plt.imshow(img)
    plt.show()
```

```python
w1 = cv2.imread('kneau.jpg')
plt.imshow(w1)
plt.show()
```

```python
face_detect(w1)
```

```python
w2 = cv2.imread('face2.jpg')
w2 = cv2.cvtColor(w2,cv2.COLOR_BGR2RGB)
plt.imshow(w2)
plt.show()
```

```python
face_detect(w2)
```

### Eye-Detection

```python
def eye_detection(img):
    eyes = eye_cascade.detectMultiScale(img)
    for ex,ey,ew,eh in eyes:
        img = cv2.rectangle(img,(ex,ey),(ex+ew,ey+eh),(0,255,0),3)
    print(ex,ey,ew,eh)
    plt.imshow(img)
    plt.show()
```

```python
w1 = cv2.imread('kneau.jpg')
plt.imshow(w1)
plt.show()
```

```python
eye_detection(w1)
```

```python
w2 = cv2.imread('face2.jpg')
w2 = cv2.cvtColor(w2,cv2.COLOR_BGR2RGB)
plt.imshow(w2)
plt.show()
```

```python
eye_detection(w2)
```

### Face_eye_detect

```python
def face_eye_detect(img):
    faces = face_cascade.detectMultiScale(img)
    for x,y,w,h in faces:        
        img = cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),3)
        face_color = img[y:y+h,x:x+w]
        eyes = eye_cascade.detectMultiScale(face_color)
        for ex,ey,ew,eh in eyes:
            cv2.rectangle(face_color,(ex,ey),(ex+ew,ey+eh),(0,255,0),3)
            
    plt.imshow(img)
    plt.show()
```

```python
w1 = cv2.imread('kneau.jpg')
plt.imshow(w1)
plt.show()
```

```python
face_eye_detect(w1)
```

```python
w2 = cv2.imread('face2.jpg')
w2 = cv2.cvtColor(w2,cv2.COLOR_BGR2RGB)
plt.imshow(w2)
plt.show()
```

```python
face_eye_detect(w2)
```
